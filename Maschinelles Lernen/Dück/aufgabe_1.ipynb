```python
def gradient_descent(X, y, alpha):
    m = len(X)  # number of training examples
    theta = [1, 1]  # initial parameters
    
    for i in range(m):
        x = X[i]
        prediction = theta[0] + theta[1]*x
        error = prediction - y[i]
        theta[0] = theta[0] - alpha * error
        theta[1] = theta[1] - alpha * error * x
        print(error)
    
    return theta
```
