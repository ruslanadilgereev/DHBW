```python
# Importieren der erforderlichen Bibliotheken
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Daten erstellen
X = np.linspace(0, 2 * np.pi, 1000).reshape(-1, 1)
y = np.sin(X).reshape(-1,)

# Aufteilung in Trainings- und Testdatens√§tze


# Modell erstellen und trainieren
mlp = MLPRegressor(hidden_layer_sizes=(10, 10, 10), max_iter=10000, epsilon=0.00001,)
mlp.fit(X, y)

# Vorhersage und Evaluation
y_pred_train = mlp.predict(X)
y_pred_test = mlp.predict(X)

train_mse = mean_squared_error(y, y_pred_train)
test_mse = mean_squared_error(y, y_pred_test)

# Ergebnisse visualisieren
plt.figure()
plt.scatter(X, y, label='Test Data')
plt.scatter(X, y_pred_test, label='Train Data')

```
