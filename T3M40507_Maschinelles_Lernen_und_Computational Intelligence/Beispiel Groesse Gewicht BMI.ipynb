```python
# BMI 100 ... wie kann die BMI Funktion per Regression gelernt werden?

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

from mpl_toolkits.mplot3d import Axes3D as ax


trainingData = 'bmi_100-1.csv'

print(os.getcwd() )

os.chdir('/Users/dirkreichardt/Downloads/')  # put your own path here ...

data = pd.read_csv(trainingData, sep=";", decimal=",")    

print(data)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter3D(data['Groesse'], data['Gewicht'], data['bmi'], color='blue')
ax.set_title('Daten Groesse/Gewicht/BMI');
plt.xlabel('Groesse')
plt.ylabel('Gewicht')

plt.show()
```

```python

plt.scatter(data['Groesse'], data['bmi'], color='blue')
plt.title('Daten Groesse -> BMI');
plt.xlabel('Groesse')
plt.ylabel('BMI')

plt.show()

plt.scatter(data['Gewicht'], data['bmi'], color='blue')
plt.title('Daten Gewicht -> BMI');
plt.xlabel('Gewicht')
plt.ylabel('BMI')

plt.show()
```
